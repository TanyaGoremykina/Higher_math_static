```python
## Задача 1.
## Известно, что генеральная совокупность распределена нормально со средним квадратическим отклонением, равным 16. 
## Найти доверительный интервал для оценки математического ожидания a с надежностью 0.95, если выборочная средняя M = 80,
## а объем выборки n = 256.
## Зная среднее крадратичное отклонение генеральной совокупности будем решать через Z.
## Исходя из условия задачи доверительный интервал = 1-0,95 = 5 % (2,5 с каждой стороны)
## По таблицам найдем A (альфу) = +/- 0,975: отрицательное = - 1,96; положительное + 1,96
import numpy as np
## положительный
Z1 = 80 + (1.96*16/256**0.5)
print (Z1)
## отрицательный
Z2 = 80 - (1.96*16/256**0.5)
print (Z2)
print ([Z2, Z1])
```

```python
## Задача 2.
## В результате 10 независимых измерений некоторой величины X, выполненных с одинаковой точностью, получены опытные данные:
## 6.9, 6.1, 6.2, 6.8, 7.5, 6.3, 6.4, 6.9, 6.7, 6.1 Предполагая, что результаты измерений подчинены нормальному закону 
## распределения вероятностей, оценить истинное значение величины X при помощи доверительного интервала, покрывающего это 
## значение с доверительной вероятностью 0,95.
x = np.array([6.9, 6.1, 6.2, 6.8, 7.5, 6.3, 6.4, 6.9, 6.7, 6.1])
## среднее арифметическое
M = x.mean()
## несмещенная дисперсия
V = x.var(ddof=1)
## несмещенное отклонение
S = V**0.5
##  т.к. среднее крадратичное отклонение генеральной совокупности неизвестно будем решать через T
## по таблице найдем А(альфа)/2 = 2,5 % (или 0,025), степень 9 т.к. (10-1) Таким образом +/- 2,262
## положительный
Z1 = M + (2.262*S/10**0.5)
print (Z1)
## отрицательный
Z2 = M - (2.262*S/10**0.5)
print (Z2)
print ([Z2, Z1])
```
