```python
## Задача 1.
## Случайная непрерывная величина A имеет равномерное распределение на промежутке (200, 800].
## Найдите ее среднее значение M(X) и дисперсию D(X).
## По условиям задачи а = 200, b = 800. Среднее значение = (а +b)/2; дисперсия = ((b-a)**2)/12
import numpy as np
M_X = (200 + 800)/2 
print (M_X)
D_X = ((800 - 200)**2)/12
print (D_X)
```

```python
## Задача 4.
## Рост взрослого населения города X имеет нормальное распределение.
## Причем, средний рост равен 174 см, а среднее квадратичное отклонение равно 8 см.
## Какова вероятность того, что случайным образом выбранный взрослый человек имеет рост:
## Исходя из условий средний рост = 174, сигма = 8,
## Воспользуемся правилом трех сигм, т.е в прелах 1 сигмы (от 166 см до 182 см) в обе стороны от среднего значения 
## находиться 68 % (34 % в каждую),
## в прелах 2 сигм (от 158 см до 190 см) в обе стороны от среднего значения 95,4 % (47,7 % в каждую или 13,7 % от границы
## первой сигмы),
## в пределах 3 сигм (от 150 см до 198 см) в обе стороны 99,72 % (49.86% в каждую или 2,16% от границы второй сигмы)
## За пределами 3 трех сигм (меньше 150 см, и больше 198 см) (100-99.72)= 0.28% или 0,14 в каждую сторону от значения 
## 3 сигмы. Таким образом 
## а). больше 182 см 
H1 = 13.7 + 2.16 + 0.14
print (H1)
## б). больше 190 см
H2 = 2.16+0.14
print (H2)
## в). от 166 см до 190 см
H3 = 34 + 47.7
print (H3)
## г). от 166 см до 182 см
H4 = 34 + 34
print (H4)
## д). от 158 см до 190 см
H5 = 47.7 + 47.7
print (H5)
## е). не выше 150 см или не ниже 190 см
H6 = 0.14 + 2.16 + 0.14
print (H6)
## ё). не выше 150 см или не ниже 198 см
H7 = 0.14 + 0.14
print (H7)
## ж). ниже 166 см.
H8 = 13.7 + 2.16 + 0.14
print (H8)
```
