```python
## Задача 2.
## В первом ящике находится 8 мячей, из которых 5 - белые. Во втором ящике - 12 мячей, из которых 5 белых. 
## Из первого ящика вытаскивают случайным образом два мяча, из второго - 4. Какова вероятность того, что 3 мяча белые?
def combi(n, k):
    return int(factorial(n) / (factorial(k) * factorial(n-k)))

## 1 ящик = 0 белых, 2 ящик = 3 белых
p1 = ((combi (5, 0) * combi (3,2)) / (combi(8, 2)) * (combi (5, 3) * combi (7, 1)) / combi (12, 4))
print (p1)

## 1 ящик = 1 белых, 2 ящик = 2 белых
p2 = ((combi (5, 1) * combi (3,1)) / (combi(8, 2)) * (combi (5, 2) * combi (7, 2)) / combi (12, 4))
print (p2)

## 1 ящик = 2 белых, 2 ящик = 1 белый
p3 = ((combi (5, 2) * combi (3,0)) / (combi(8, 2)) * (combi (5, 1) * combi (7, 3)) / combi (12, 4))
print (p3)

P = p1 + p2 + p3
print (P)
```

```python
## Задача 3.
## На соревновании по биатлону один из трех спортсменов стреляет и попадает в мишень.
## Вероятность попадания для первого спортсмена равна 0.9, для второго — 0.8, для третьего — 0.6. 
## Найти вероятность того, что выстрел произведен:
## a). первым спортсменом б). вторым спортсменом в). третьим спортсменом.
## формула Байса, решаем через полную вероятность
P = (1/3*0.9) + (1/3*0.8) + (1/3*0.6)
print (P)

## вероятность того, что выстрел произведен первым спортсменом
p1 = 1/3*0.9/P
print (p1)

## вероятность того, что выстрел произведен вторым спортсменом
p2 = 1/3*0.8/P
print (p2)

## вероятность того, что выстрел произведен третьим спортсменом
p3 = 1/3*0.6/P
print (p3)

## проверка
p = p1 + p2 + p3
print (p)
```
